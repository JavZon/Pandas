```python
import pandas as pd
import numpy as np
```

```python
dict1={
    'UZB':[32.9,32.5,31.9,31.4],
    'KAZ':[18.8,18.5,18.1,17.8],
    'TAJ':[9.5, 9.3, 9.1, 8.7],
    'KYR':[7.5, 7.2, 6.8, 6.6]
}
df=pd.DataFrame(dict1, index=[2021,2020,2019,2018])
df
```

```python
df.prod(axis=1)
```

```python
def funqsiya(x):
  return 'High' if x>32 else 'Low'
df['AFG']=df['UZB'].apply(funqsiya)
df
```

```python
df['TUR']=df['KYR'].apply(lambda x : 'BOY' if x> 7 else "QAMBAG'AL")
df
```

```python
df['IRN']=df.apply(lambda x : x['UZB'] + x['KAZ'], axis=1)
df
```
