```python
import pandas as pd
import numpy as np

www = {
    'A': [[25, 96], [], [97, 21, 30], [13], [67, 33]]
}

df = pd.DataFrame(www)
```

```python
def split_values(row):
    if len(row) >= 2:
        return row[0], row[1], np.nan
    elif len(row) == 1:
        return row[0], np.nan, np.nan
    else:
        return np.nan, np.nan, np.nan
```

```python
df[['Col1', 'Col2', 'Col3']] = df['A'].apply(lambda x: pd.Series(split_values(x)))
```
