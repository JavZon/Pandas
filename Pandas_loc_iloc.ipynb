```python
import pandas as pd

data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data, index=['N1', 'N2', 'N3'])
# Single label
print(df.loc['N1', 'A'])  # Access a single cell
# List of labels
print(df.loc[['N1', 'N2'], ['A', 'B']])  # Access specific rows and columns
# Slice of labels
print(df.loc['N1':'N2', 'A':'B'])  # Access a range of rows and columns
# Boolean array
condition = df['B'] > 4
print(df.loc[condition, ['A', 'C']])  # Filter rows based on a condition
# Callable function
print(df.loc[lambda df: df['A'] > 2])  # Access rows based on a custom condition
```
