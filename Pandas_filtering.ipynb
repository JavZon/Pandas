```python
import pandas as pd
import numpy as np

# Create a DataFrame
data = {
    'Alex': [3,1,1,4],
    'Bob': [6,5,9,5],
    'Charlie': [4,9,7,5],
    'Daniel': [6,8,6,4],
    'Elizabeth': [4,7,3,0]
}
df = pd.DataFrame(data, index=['row1', 'row2', 'row3', 'row4'])
```

```python
df
```

```python
result = df[df['Alex'] > 5]

#This filters the rows where values in column 'A' are less than 3.
result
```

```python
result = df[df['Alex'] < 3]
result
```

```python
res=df[(df['Alex']>3) & (df['Bob']<6)]
res
```

```python
result = df[(df['Alex'] > 3) & (df['Bob'] < 5)]
result
```

```python
result = df[df['Bob'].isin([2, 4, 6])]
result
```

```python
result = df.loc[df['Alex'] < 3, ['Bob', 'Charlie']]
result
```
