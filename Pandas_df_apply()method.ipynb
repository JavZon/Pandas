```python
import pandas as pd

def subtract_row_average(df):
    # Calculate row-wise mean
    row_means = df.mean(axis=1)

    # Subtract row-wise mean from each element in the corresponding row
    result_df = df.sub(row_means, axis=0)

    return result_df

# Example usage:
# Assuming your DataFrame is named 'my_dataframe'
# Replace 'my_dataframe' with the actual name of your DataFrame
my_dataframe = pd.DataFrame({
    'A': [10, 4, 5, 2],
    'B': [7, 2, 8, 3],
    'C': [6, 9, 12, 15],
    'D': [3, 1, 4, 2]
}, index=['r1', 'r2', 'r3', 'r4'])

result_dataframe = subtract_row_average(my_dataframe)

# Display the result
print(result_dataframe)
```

```python
import pandas as pd
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
double_function = lambda x: x*2
column_doubles_apply = df.apply(double_function)
element_doubles_applymap = df.applymap(double_function)
# Using df.apply() to double each row
row_doubles_apply = df.apply(double_function, axis=1)
# Displaying the results
print("df.apply():\n", column_doubles_apply)
print("\ndf.applymap():\n", element_doubles_applymap)
print("\ndf.apply() along rows:\n", row_doubles_apply)
```
