```python
data2={
    'Alex':[78,41,90,512,98,90,51,67],
    'Bob':[12,14,0,12,8,9,5,672],
    'Charlie':[33,1,90,2,8,14,521,367],
    'David':[44,41,95,52,18,94,51,67]
    }
index=['r1','r2','r3','r4','r5','r6','r7','r8',]
df2=pd.DataFrame(data2, index=index)
df2
```

```python
df2.sort_index(axis=1)
```

```python
df2.mean(axis=1)
```

```python
def row_means (df):
  return df.apply(lambda x: x.mean(), axis=1 )

df2.sub(row_means(df2), axis='index')
```
