```python
import pandas as pd
import numpy as np

# Create a DataFrame with missing values
data = {'A': [1, 2, np.nan, 4, np.nan], 'B': [5, np.nan, 7, np.nan, 9]}
df = pd.DataFrame(data)

# Example 1: Reindex with forward fill method
new_index = [0, 1, 2, 3, 4]
df_reindexed_ffill = df.reindex(new_index, method='ffill')


# Example 2: Reindex with backward fill method
new_index = [0, 1, 2, 3, 4]
df_reindexed_bfill = df.reindex(new_index, method='bfill')
print("\nExample 2 - Backward Fill:")
print(df_reindexed_bfill)

# Example 3: Reindex with interpolate method
new_index = [0, 1, 2, 3, 4]
df_reindexed_interpolate = df.reindex(new_index)
df_reindexed_interpolate = df_reindexed_interpolate.interpolate()
print("\nExample 3 - Interpolate:")
print(df_reindexed_interpolate)

# Example 4: Reindex with specific fill values
new_index = [0, 1, 2, 3, 4]
fill_values = {'A': 0, 'B': 100}
df_reindexed_fill_values = df.reindex(new_index).fillna(fill_values)
print("\nExample 4 - Fill with Custom Values:")
print(df_reindexed_fill_values)
```

```python
import pandas as pd
import numpy as np

data = {'A': [1, 2, np.nan, 4, 5], 'B': [6, np.nan, 8, np.nan, 10]}
df = pd.DataFrame(data)

# Fill NaN with a specific value (e.g., 0)
df_filled = df.fillna(0)

# Forward fill NaN values in the DataFrame
df_ffill = df.fillna(method='ffill')

# Backward fill NaN values in the DataFrame
df_bfill = df.fillna(method='bfill')

# Fill NaN values with a dictionary specifying different values for each column
fill_values = {'A': 99, 'B': 100}
df_dict_filled = df.fillna(value=fill_values)
```
