```python
import pandas as pd
```

```python
data={
    'Year':[2020,2018,2023,2021],
    'Mileage':[1200, 3100, 500, 900],
    'Price':[15000, 20000, 13000, 9000]
}
ind=['Cobalt','Malibu','Gentra','Spark']

df1=pd.DataFrame(data, index=ind)
df1
```

```python
df1.rank()
```

```python
data2={
    'Fruit':['Banana','Apple','Apple','Banana','Banana','Apple'],
    'Price':[2500,3100,1700,2250,3600,4000],
    'Expiry_date':[10,7,14,9,6,8]
}
df2=pd.DataFrame(data2)
```
