```python
import pandas as pd

data = {
    'Age': [20, 35, 27, 41],
    'Name': ['John', 'Abu', 'Tod', 'Ali'],
    'Job': ['Doctor', 'Teacher', 'Lawyer', 'Policeman'],
    'Salary': [None, None, None, None]
}

df = pd.DataFrame(data, index=[1, 2, 3, 4])

df
```

```python
df['Salary']=[15000,27000,33000,25000]
df
```

```python
df['Living Standart']=['Rich' if x>25000 else 'Poor' for x in df['Salary'] ]
df
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Age', y='Salary', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
